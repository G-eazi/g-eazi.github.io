#Counting Overlaps code for ORI Recognition below

```python
def PatternCount(Text,Pattern):
    count=0
    for i in range(len(Text)-len(Pattern)+1):#i is all the possible start values of the string,2=0,1 in python. It tests the number a
        if Text[i:i+len(Pattern)]==Pattern:
            count+=1
    return count
```

#Frequency Table Code Below(Dictionary)

```python
def FrequencyMap(Text,k):
    freqMap={}
    n=len(Text)
    for i in range(n-k+1): #addding one gets all possible values cause in python count from 0 to 1
        Pattern=Text[i:i+k]#k is the pattern number (k=3=AAA, K=4=TTAT )
        if Pattern in freqMap:
            freqMap[Pattern]+=1
        else:
            freqMap[Pattern]=1
    return freqMap
```

```python
def count_mismatch(seq1,seq2):
  count=0
  if len(seq1) != len(seq2):
    print("length of sequences not matching")
  else:
    for i in range(len(seq1)):
      if seq1[i] != seq2[i]:
        count+=1
  return count
```

```python
def FrequentStringMisMatch(Pattern,Genome,m):
    position=[]
    for i in range(len(Genome)-len(Pattern)+1):
        seq2=Genome[i:i+len(Pattern)]
        if count_mismatch(Pattern,seq2) <= m:
            position.append(i)
    return position
```

```python
#Code gets the highest frequency pattern for a given length in the DNA string from Freq Map dict then Pattern in a list.
```

```python
def BetterFrequentWords(Text, k):
  FrequentPatterns=[]
  freqMap=FrequencyMap(Text,k)
  maxM = max(freqMap.values()) #returns output/count and the max picks the highest
  for Pattern in freqMap:
    if freqMap[Pattern] == maxM:
      FrequentPatterns.append(Pattern)
  return FrequentPatterns
```

```python
def FindClumps(Text,k,L,t): #How many times a kmer is seen in a window of genome(piece of genome) for the times we want to see it
  Patterns=[]
  n=len(Text)
  for i in range(n-L+1):#starting positions for string
    Window=Text[i:i+L] #generates kmers based on starting positions and L
    freqMap=FrequencyMap(Window,k) #we get the occurences of a given kmer
    for s in freqMap: #we compare the occurences of the kmer to the times we expect to see it.
        if freqMap[s] >= t and s not in Patterns:  #if count/occurrences from dict is > or = times we expect from clump/window of DNA string add it to Patterns list
            Patterns.append(s) #not in remove duplicates from Patterns and add it
  return Patterns
```

```python
def generate_neighbors(pattern, d):
    if d == 0:
        return [pattern]
    neighbors = set()
    for i in range(len(pattern)):
        for nucleotide in 'ACGT':
            if nucleotide != pattern[i]:
                new_pattern = pattern[:i] + nucleotide + pattern[i+1:]
                neighbors.update(generate_neighbors(new_pattern, d - 1))
    return list(neighbors)

    return generate_neighbors(Pattern, d)
```

```python
def FrequentWordsWithMismatches(Text, k, d):
    """Finds the most frequent k-mers with at most d mismatches in a given text."""
    Mismatch_Patterns = []  # To store the most frequent k-mers with mismatches.
    FreqMismatchMap = {}  # A dictionary to store the frequency of each k-mer (with mismatches).
    n = len(Text)

    # Loop through all substrings of length k.
    for i in range(n - k + 1):
        pattern = Text[i:i + k]  # Extract the current k-mer.
        # Get the neighborhood (k-mers with mismatches).
        neighborhood = generate_neighbors(pattern, d)
        for mismatch in neighborhood:
            if mismatch in FreqMismatchMap:
                FreqMismatchMap[mismatch] += 1  # Increment the frequency of the mismatch pattern.
            else:
                FreqMismatchMap[mismatch] = 1  # Initialize the frequency for the mismatch pattern.

    # Find the maximum frequency.
    maxM = max(FreqMismatchMap.values())

    # Collect patterns that have the maximum frequency.
    for dna_s in FreqMismatchMap:
        if FreqMismatchMap[dna_s] == maxM:
            Mismatch_Patterns.append(dna_s)

    return Mismatch_Patterns  # Return the list of most frequent k-mers with mismatches.
```

```python
def frequent_words_with_mismatches(text, k, d):
    patterns = {}
    for i in range(len(text) - k + 1):
        pattern = text[i:i+k]
        neighborhood = generate_neighbors(pattern, d)
        for neighbor in neighborhood:
            patterns[neighbor] = patterns.get(neighbor, 0) + 1

    max_count = max(patterns.values())
    return [pattern for pattern, count in patterns.items() if count == max_count]
```
